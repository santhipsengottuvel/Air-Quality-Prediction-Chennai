### Splitting the dataset by station name and storing in the variable of Station name

```python
a=['Adyar']
b=['Anna Nagar']
c=['Thiyagaraya Nagar']
d=['Kilpauk']
e=['Nungampakkam']
f=['Kathiwakkam']
g=['Manali']
h=['Tiruvottiyur']

Adyar=df[df['Stations'].isin(a)].reset_index()
Anna_Nagar=df[df['Stations'].isin(b)].reset_index()
Thiyagaraya_Nagar=df[df['Stations'].isin(c)].reset_index()
Kilpauk=df[df['Stations'].isin(d)].reset_index()
Nungampakkam=df[df['Stations'].isin(e)].reset_index()
Kathiwakkam=df[df['Stations'].isin(f)].reset_index()
Manali=df[df['Stations'].isin(g)].reset_index()
Tiruvottiyur=df[df['Stations'].isin(h)].reset_index()
```

## Analysis between the stations by Total average concentrations of pollutants

```python
df.head()
```

```python
# Finding the pollutants average separately based on the stations by grouping
Sulphur=df.groupby("Stations")[["Sulphur_dioxide"]].mean() # gives sulphur dioxide average for every stations seperately
Nitrogen=df.groupby("Stations")[["Nitrogen_dioxide"]].mean() # gives nitrogen dioxide average for every stations seperately
Rspm=df.groupby("Stations")[["PM_10"]].mean() # gives PM 10 average for every stations seperately
PM=df.groupby("Stations")[["PM_2.5"]].mean() # gives PM 2.5 average for every stations seperately
```

```python
Sulphur.head()
```

```python
# Concatenating all the pollutants in a single data frame and storing them in variable
New=pd.concat([Sulphur,Nitrogen,Rspm,PM],axis=1,join="inner")
New
```

```python
sns.heatmap(New,annot=True,cmap="Spectral")
```

```python
New["Sulphur_dioxide"].plot(kind=("bar"),figsize=(16,6),xlabel="Stations",ylabel="Concentration",title="Sulphur dioxide Concentration",fontsize=12)
```

From this bar chart it is well shown that Manali, Kathiwakkam, Tiruvottiyur experiences high concentration of Sulphur dioxide because they are all the industrial areas. The main cause of sulphur dioxide emission is industrial emission so that these 3 stations experiences high concentration of SO2 than other stations

```python
New["Nitrogen_dioxide"].plot(kind=("bar"),figsize=(16,6),xlabel="Stations",ylabel="Concentration",title="Nitrogen dioxide Concentration",fontsize=12)
```

From this bar chart it is well shown that Thiyagaraya Nagar experiences high concentration of Nitrogen dioxide than other stations. Behind Thiyagaraya Nagar, Anna Nagar and Nungampakkam faced high concentration level of Nitrogen dioxide.

```python
New["PM_10"].plot(kind=("bar"),figsize=(16,6),xlabel="Stations",ylabel="Concentration",title="PM10 Concentration",fontsize=12)
```

From this bar chart it is well shown that Anna Nagar experiences high concentration of PM 10 than other stations. Behind Anna Nagar, Thiyagaraya Nagar and Nungampakkam faced high concentration level of PM 10.

```python
New["PM_2.5"].plot(kind=("bar"),figsize=(16,6),xlabel="Stations",ylabel="Concentration",title="PM2.5 Concentration",fontsize=12)
```

From this bar chart it is well shown that Manali experiences high concentration of PM 2.5 than other stations. Behind Manali, Thiruvottiyur and Kathiwakkam faced high concentration level of PM 2.5.

```python
New.plot(kind="bar",figsize=(17,8),xlabel="Stations",ylabel="Concentration",)
```

This Bar graph shows that PM!0 is the major pollutant in chennai city and sulphur dioxide is minor Pollutant. This graph also indicates their levels in all stations
